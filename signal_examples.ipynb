```python
import numpy as np
from math import pi
%matplotlib inline
from matplotlib import pyplot as plt
np.random.seed(0)
from scipy import signal as sig
```

```python
plt.rcParams["figure.figsize"] = (18, 6) # (w, h)
def plot(signal, siglim=0, ylim=0):
    plt.subplot(1,2,1)
    if siglim != 0:
        plt.ylim(siglim)
    else:
        plt.ylim(0, pi)
    plt.xlim(0,300)
    plt.plot(signal)
    plt.subplot(1,2,2)
    if ylim != 0:
        plt.ylim(0, ylim)
    plt.xlim(0,300)
    fft = np.abs(np.fft.fft(signal))
    fft[0] = fft[1]
    #fft = fft[:150]
    plt.plot(fft)
```

```python
noise = np.sin(2*pi*np.arange(300)/13)*0.1
plot(noise, siglim=(-0.2, 0.2))
```

```python
sinus = np.sin(np.arange(300)/23*pi)*pi + pi + noise
plot(sinus, siglim=(0, 2*pi))
```

```python
random_vals = np.random.random_sample(300)
random_vals = np.multiply(pi, random_vals)
random_vals = random_vals + noise
plot(random_vals)
```

```python
def random_movement(strength):
    movement = np.array([pi/2])
    while len(movement) < 300:
        last_val = movement[-1]
        new_val = last_val + np.random.rand(1)*strength - strength/2
        if new_val > pi:
            new_val = pi
        elif new_val < 0:
            new_val = 0
        movement = np.append(movement, [new_val])
    return movement
```

```python
movement = random_movement(0.2) + noise
plot(movement, ylim=35)
```
